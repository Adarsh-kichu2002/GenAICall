```python
import re
import os

# Check current working directory
print(f"Current working directory: {os.getcwd()}")

# List files in current directory
print(f"Files available: {os.listdir('.')}")

# 1. LOAD: Read the file from Step 1
try:
    with open("1_raw_transcript.txt", "r") as f:
        raw_text = f.read()
    print(f"Successfully loaded 1_raw_transcript.txt ({len(raw_text)} characters)")
except FileNotFoundError as e:
    print(f"Error: {e}")
    raw_text = None

if raw_text:
    # 2. PROCESS: Your cleaning function
    def clean_text(text):
        text = text.lower()
        text = re.sub(r'\b(um|uh|hmm|ah|like)\b', '', text)
        return re.sub(r'\s+', ' ', text).strip()

    cleaned_text = clean_text(raw_text)

    # 3. SAVE: Write to a new file for the next script
    with open("2_cleaned_transcript.txt", "w") as f:
        f.write(cleaned_text)

    print("Step 2 Complete: 2_cleaned_transcript.txt created.")
```
